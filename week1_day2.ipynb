```python
grades = [45,67,90,78]

#الوصول الى العناصر في القائمة 
print("العنصر الاول في القائمة",grades[0])#العنصر الاول في القائمة
print("العنصر الرابع في القائمة",grades[3])#العنصر الرابع في القائمة
```

```python
grades = [45,67,90,78]

#اضافة عنصر الى القائمة
grades.append(56)
print(grades)
```

```python
student = ['ahmed', 'karim', 'jawad', 'sara']

#استبدال عنصر في القائمة
student[0]= 'hajar'
print(student)

#المرور على الثلاميد في القائمة 
for i in student:
    print(i)    

#الاستفسار عن عنصر في القائمة
'ahmed' in student
```
