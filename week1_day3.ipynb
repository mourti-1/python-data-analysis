```python
sales = []  #قائمة فارغة

for i in range(4):  #المرور على كل عنصر في مجال 2
    data_days = int(input(f"day {i+1} sales: "))  #ادخال بيانات يومية و تحويلها الى ععد
    sales.append(data_days)  #تخزين بيانات الادخال في القائمة الفارغة

total = sum(sales)  #حساب مجموع العناصر في القائمة
print("Total: ",total)

avg = total / len(sales)  #حساب متوسط العناصر في القائمة
print('average: ',avg)

top_sales = max(sales)  #اعلى المبيعات او اكبر رقم في القائمة
lowest_sales = min(sales)  #ادنى المبيعات
print('Top sales: ',top_sales)
print('Lowest sales: ',lowest_sales)

best_day = sales.index(top_sales)  #يوم اعلى المبيعات او اعلى رقم في القائمة
worst_day = sales.index(lowest_sales)  #اسوء يوم في المبيعات
print('Best day: ',best_day +1)
print('Worst day: ',worst_day +1)#زياد رقم 1 لتعطي اليوم الحقيقي من القائمة

#مقارنة بين المتوسط و المبيعات في القائمة
for x in range(len(sales)):  #المرور على كل عنصر من مجال القائمة
    if sales[x] > avg :  #شرط المقارنة بين متوسط المبيعات و مبيعات الايام في القائمة 
        print('Day above average:',x+1)  #طباعة رقم اليوم

     #تصنيف البيانات
print('Sales Classification: ')
for y in range(len(sales)):
    if sales[y] / avg >= 1.2 :
        print('Day excellent: ',y+1)
    elif sales[y] > avg:
        print('Day good: ',y+1)
    elif sales[y] == avg:
        print('Day average: ',y+1)
    else :
        print('Day low: ',y+1)
```
