```python
sales = [100, 150, 80, 200, 170]
#مجموع عدد المبيعات 
Totale_sales = sum(sales)
print("Sales:",Totale_sales)
#عدد الايام
Days = len(sales)
print("days:",Days)
#متوسط المبيعات
avg = Totale_sales / Days
print("average:",avg)
Max_sales = max(sales)#اعلى المبيعات 
Min_sales = min(sales)#ادنى المبيعات
print("Highest sales:",Max_sales)
print("Lowest sales:",Min_sales)
Top_Sales_Day = sales.index(Max_sales) #موضع اكبر قيمة في القائمة
print("Top Sales Day:",Top_Sales_Day + 1)

```

```python
sales = []
for i  in range(3):
    data_in = int(input(f"day {i+1} sales:"))
    sales.append(data_in)
    
#المجموع
total = sum(sales)
print("Totale:",total)

#المتوسط
avg = total / len(sales)
print("average :",avg)

#اعلى المبيعات
Max_sales = max(sales)
print("Top sales:",Max_sales)

#احسن يوم في المبيعات
best_day = sales.index(Max_sales)
print("best day:",best_day+1)
#هل هناك ايام كانت اكبر من متوسط المبيعات
for x in range(len(sales)):#لنفحص جميع الايام في المبيعات 
    if sales[x] > avg:#اذا كان مكان  اكس اكبر من متوسط المبيعات
        print("Days above average:",x+1)#اطبع رقم اليوم زائد واحد
```
